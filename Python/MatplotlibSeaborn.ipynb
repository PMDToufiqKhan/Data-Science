## MATPLOTLIB & SEABORN COMPREHENSIVE REFERENCE

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
## Create sample data
np.random.seed(42)
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Age': [25, 30, 35, 40, 28, 33],
    'Salary': [50000, 60000, 75000, 80000, 55000, 70000],
    'Dept': ['HR', 'IT', 'IT', 'Finance', 'HR', 'Finance'],
    'Years': [2, 5, 8, 10, 3, 6]
})
df
```

### 1. MATPLOTLIB BASICS

```python
## Create figure and axes
fig, ax = plt.subplots()                    # single plot
```

```python
fig, axes = plt.subplots(2, 2)              # 2×2 grid (returns array)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # set size (width, height)
```

```python
## Figure attributes
fig, ax = plt.subplots()
fig.suptitle('Main Title')                  # title for entire figure
fig.set_size_inches(12, 8)                  # change figure size
fig.tight_layout()                          # auto-adjust spacing
plt.show()
```

### 2. BASIC LINE PLOT

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y)                              # basic line plot

plt.plot(x, y, label='Series 1')            # add label

plt.plot(x, y, color='red')                 # set color

plt.plot(x, y, linestyle='--')              # dashed line (-, --, -., :)

plt.plot(x, y, linewidth=2)                 # line thickness

plt.plot(x, y, marker='o')                  # add markers (o, s, ^, *, +, x)

plt.plot(x, y, marker='o', markersize=8)    # marker size

plt.xlabel('X-axis')                        # x label

plt.ylabel('Y-axis')                        # y label

plt.title('Line Plot')                      # plot title

plt.legend()                                # show legend

plt.grid(True)                              # show grid
plt.show()                                  # display plot
```

### 3. SCATTER PLOT

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Age'], df['Salary'])        # basic scatter
plt.scatter(df['Age'], df['Salary'], 
            c=df['Years'],                  # color by another variable
            s=100,                          # marker size
            alpha=0.6,                      # transparency (0-1)
            cmap='viridis',                 # colormap
            edgecolors='black',             # edge color
            linewidth=1)                    # edge width
plt.colorbar(label='Years')                 # color scale legend
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Scatter: Age vs Salary')
plt.show()
```

### 4. BAR PLOT

```python
plt.figure(figsize=(8, 5))
plt.bar(df['Name'], df['Salary'])           # vertical bar
plt.bar(df['Name'], df['Salary'], 
        color=['red', 'blue', 'green', 'orange', 'purple', 'brown'])
plt.xticks(rotation=45)                     # rotate x labels
plt.xlabel('Name')
plt.ylabel('Salary')
plt.title('Salary by Person')
plt.show()
```

```python
## Horizontal bar plot
plt.barh(df['Name'], df['Salary'])          # horizontal bar
plt.xlabel('Salary')
plt.ylabel('Name')
plt.title('Salary by Person (Horizontal)')
plt.show()
```

```python
## Grouped bar plot
x_pos = np.arange(len(df['Name']))
width = 0.35
plt.figure(figsize=(10, 5))
plt.bar(x_pos - width/2, df['Age'], width, label='Age')
plt.bar(x_pos + width/2, df['Years'], width, label='Years')
plt.xticks(x_pos, df['Name'], rotation=45)
plt.legend()
plt.title('Grouped Bar Plot')
plt.show()
```

### 5. HISTOGRAM

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Age'])                         # basic histogram
plt.hist(df['Age'], bins=10)                # number of bins
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
## Multiple histograms
plt.figure(figsize=(8, 5))
plt.hist(df['Age'], label='Age', alpha=0.7)
plt.hist(df['Years'], label='Years', alpha=0.7)
plt.legend()
plt.title('Multiple Histograms')
plt.show()
```

### 6. BOX PLOT

```python
plt.figure(figsize=(8, 5))
plt.boxplot(df['Salary'])                   # basic boxplot
plt.ylabel('Salary')
plt.title('Salary Distribution')
plt.show()
```

```python
## Multiple boxplots
plt.figure(figsize=(8, 5))
plt.boxplot([df[df['Dept']=='HR']['Salary'],
             df[df['Dept']=='IT']['Salary'],
             df[df['Dept']=='Finance']['Salary']],
            tick_labels=['HR', 'IT', 'Finance'])
plt.ylabel('Salary')
plt.title('Salary by Department')
plt.show()
```

### 7. SUBPLOTS (Multiple plots in one figure)

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].plot(df['Age'], df['Salary'])    # top-left
axes[0, 0].set_title('Line Plot')

axes[0, 1].scatter(df['Age'], df['Salary'])  # top-right
axes[0, 1].set_title('Scatter Plot')

axes[1, 0].bar(df['Name'], df['Salary'])     # bottom-left
axes[1, 0].set_title('Bar Plot')
axes[1, 0].tick_params(axis='x', rotation=45)

axes[1, 1].hist(df['Age'], bins=5)           # bottom-right
axes[1, 1].set_title('Histogram')

fig.suptitle('Multiple Subplots', fontsize=16)
plt.tight_layout()
plt.show()
```

### 8. CUSTOMIZATION – COLORS, STYLES, FONTS

```python
## Colors
# Named: 'red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray'
# Hex: '#FF5733'
# RGB: (0.5, 0.2, 0.8)
```

```python
## Line styles
# '-' solid line, '--' dashed, '-.' dash-dot, ':' dotted
```

```python
## Markers
# 'o' circle, 's' square, '^' triangle, '*' star, '+' plus, 'x' X, 'D' diamond
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Age'], df['Salary'], 
         color='#FF5733',                   # hex color
         linestyle='--',                    # dashed
         linewidth=3,                       # thick
         marker='o',                        # circle markers
         markersize=10,
         label='Data')

plt.xlabel('Age', fontsize=12, fontweight='bold')
plt.ylabel('Salary', fontsize=12, fontweight='bold')
plt.title('Customized Plot', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)                   # semi-transparent grid
plt.tight_layout()
plt.show()
```

### 10. SEABORN PLOTS

```python
## Regression plot (with trend line)
plt.figure(figsize=(8, 5))
sns.regplot(x='Age', y='Salary', data=df)   # scatter + regression line
plt.title('Regression Plot')
plt.show()
```

```python
## Line plot (better than matplotlib for DataFrames)
plt.figure(figsize=(8, 5))
sns.lineplot(x='Age', y='Salary', data=df)
plt.title('Line Plot')
plt.show()
```

```python
## Scatter plot (seaborn version)
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='Salary', data=df, hue='Dept', s=200)
plt.title('Scatter with Hue')
plt.show()
```

```python
## Bar plot (seaborn version)
plt.figure(figsize=(8, 5))
sns.barplot(x='Dept', y='Salary', data=df)  # mean salary per dept
plt.title('Bar Plot')
plt.show()
```

```python
## Count plot (frequency)
plt.figure(figsize=(8, 5))
sns.countplot(x='Dept', data=df)            # count per category
plt.title('Count Plot')
plt.show()
```

```python
## Box plot (seaborn version)
plt.figure(figsize=(8, 5))
sns.boxplot(x='Dept', y='Salary', data=df)  # boxplot per department
plt.title('Box Plot')
plt.show()
```

```python
## Violin plot (distribution shape)
plt.figure(figsize=(8, 5))
sns.violinplot(x='Dept', y='Salary', data=df)  # distribution + box
plt.title('Violin Plot')
plt.show()
```

```python
## Histogram with KDE (kernel density)
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Age', kde=True, bins=10)
plt.title('Histogram with KDE')
plt.show()
```

```python
## Heatmap (correlation matrix)
plt.figure(figsize=(8, 5))
corr = df[['Age', 'Salary', 'Years']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
## Pair plot (all pairwise relationships)
sns.pairplot(df[['Age', 'Salary', 'Years', 'Dept']], hue='Dept')
plt.show()
```

```python
## Distribution plot (univariate distribution)
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Salary', kde=True)
plt.title('Distribution Plot')
plt.show()
```

### 11. FACETING / MULTI-PANEL PLOTS

```python
## Using FacetGrid
g = sns.FacetGrid(df, col='Dept', height=4, aspect=1)
g.map(plt.scatter, 'Age', 'Salary')
plt.show()
```

### 12. CATEGORICAL PLOTS

```python
## Strip plot (scatter with jitter)
plt.figure(figsize=(8, 5))
sns.stripplot(x='Dept', y='Salary', data=df, size=8)
plt.title('Strip Plot')
plt.show()
```

```python
## Swarm plot (no overlap)
plt.figure(figsize=(8, 5))
sns.swarmplot(x='Dept', y='Salary', data=df, size=8)
plt.title('Swarm Plot')
plt.show()
```

### 13. JOINT PLOTS (Bivariate + Marginal distributions)

```python
sns.jointplot(x='Age', y='Salary', data=df, kind='scatter')
plt.show()
```

```python
sns.jointplot(x='Age', y='Salary', data=df, kind='hex')  # hexbin
plt.show()
```

```python
sns.jointplot(x='Age', y='Salary', data=df, kind='kde')  # density
plt.show()
```

### 17. PIE CHART

```python
plt.figure(figsize=(8, 5))
dept_counts = df['Dept'].value_counts()
plt.pie(dept_counts, labels=dept_counts.index, autopct='%1.1f%%')
plt.title('Department Distribution')
plt.show()
```

### 18. 3D PLOTS (optional)

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Age'], df['Salary'], df['Years'])
ax.set_xlabel('Age')
ax.set_ylabel('Salary')
ax.set_zlabel('Years')
plt.show()
```

### 19. ANNOTATE / TEXT ON PLOTS

```python
fig, ax = plt.subplots()
ax.scatter(df['Age'], df['Salary'])
for i, name in enumerate(df['Name']):
    ax.annotate(name, (df['Age'].iloc[i], df['Salary'].iloc[i]))
ax.set_xlabel('Age')
ax.set_ylabel('Salary')
plt.show()
```
