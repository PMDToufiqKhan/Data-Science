```python
# Correlation ranges from -1 to +1:

# Value	Meaning
# +1.0	Perfect positive correlation (both increase together)
# +0.5 to +0.9	Strong positive correlation
# +0.0 to +0.5	Weak positive correlation
# 0.0	No correlation (independent)
# -0.5 to -0.9	Strong negative correlation
# -1.0	Perfect negative correlation (one increases, other decreases)
```

```python
df['Salary'].skew()           # skewness
```

```python
# Value	        Name	                  Distribution Shape	       Meaning
# 0	            Symmetric	              Bell curve, centered	       Data is perfectly balanced
# +0.5 to +1	Moderately right-skewed	  Tail on right	               Few high values, most low
# > +1	        Highly right-skewed	      Long tail right	           Very few high outliers
# -0.5 to -1	Moderately left-skewed	  Tail on left	               Few low values, most high
# < -1	        Highly left-skewed	      Long tail left	           Very few low outliers
```

```python
df['Salary'].kurtosis()       # kurtosis
                              # Kurtosis measures how PEAKED the data is and how FAT the tails are.
```

```python
# Range	          Category	              Meaning
# -0.5 to +0.5	  Very close to 0	      Nearly normal, almost no skew from standard
# +0.5 to +1.0	  Slightly positive	      Slightly more peaked, few more outliers
# +1.0 to +2.0	  Moderately positive	  Noticeably peaked, more outliers expected
# > +2.0	      High positive	          Very peaked, many outliers
# -0.5 to -1.0	  Slightly negative	      Slightly flatter, fewer outliers
# -1.0 to -2.0	  Moderately negative	  Noticeably flat, very few outliers
# < -2.0	      High negative	          Very flat, almost no outliers
```

#### 10. MERGING / JOINING / COMBINING

```python
df1 = pd.DataFrame({
    'Key' : [1, 2, 3, 4, 5],
    'Value' : [34, 65, 63, 39, 75]
})
df2 = pd.DataFrame({
    'Key' : [0, 3, 6, 5, 8],
    'Value' : [85, 15, 87, 19, 64]
})
```

```python
pd.merge(df1, df2, on='Key')          # inner join on 'Key'
```

```python
pd.merge(df1, df2, on='Key', how='left')   # left join
```

```python
pd.merge(df1, df2, on='Key', how='right')  # right join
```

```python
pd.merge(df1, df2, on='Key', how='outer')  # outer join
```

```python
pd.concat([df1, df2])                 # stack vertically
```

```python
pd.concat([df1, df2], axis=1)         # stack horizontally
```

#### 18. ITERATION (usually avoid; vectorize instead)

```python
for idx, row in df.iterrows():        # iterate by row (slow)
    print(row['Name'], row['Age'])
```

```python
for row in df.itertuples(index=True):      # faster iteration
    print(row.Name, row.Age)
    print(row.Index)  # access index via row.Index
```

```python
for col, series in df.items():        # iterate by column
    if series.dtype in ['int64', 'float64']:  # only numeric
        print(col, series.mean())
```
